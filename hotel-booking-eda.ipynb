# Importing libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

# Loading Dataset

```python
df=pd.read_csv("/kaggle/input/hotel-booking/hotel_booking.csv")
```

# Exploratory Data Analysis

```python
# dropping undesired columns
df.drop(['name', 'email',
       'phone-number', 'credit_card'],axis=1 ,inplace=True)
```

```python
df.info()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.info()
```

```python
df.describe(include='object')
```

```python
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
df.isnull().sum()
```

```python
# too much missing info in agent and company column, hence dropping these column

df.drop(['agent','company'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# box plot to check outlier in adr column
df['adr'].plot(kind='box')
df = df[df['adr']<5000] # extreme outlier
```

# Data Analysis and Visualizations

```python
df.head()
```

```python
cancelled_prec = df['is_canceled'].value_counts(normalize=True)
print(cancelled_prec)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not_Cancelled','Cancelled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled', data=df, palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[1,1])
plt.title("Reservation status in different hotels", size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservations')
plt.show()
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel = df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label='Resort_hotel')
plt.plot(city_hotel.index, city_hotel['adr'],label='City_hotel')
plt.legend()
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x = 'month', data=df, hue='is_canceled')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[1,1])
plt.title("Reservation status per month", size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month',y='adr', data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
# day of the week vs cancellations
df['day_of_week']=df.reservation_status_date.apply(lambda i: i.day_name())
```

```python
plt.figure(figsize=(12,5))
order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
ax1=sns.countplot(x = 'day_of_week', data=df, hue='is_canceled', order=order)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[1,1])
plt.title("Reservation status per day of week", size=20)
plt.xlabel('Day of week')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.show()
```

```python
# comparing cancellation on weekend or weekday for city or resort hotels.

city_hotel_df = df[df['hotel']=='City Hotel']
resort_hotel_df = df[df['hotel']=='Resort Hotel']

order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
plt.figure(figsize=(18,5))
plt.subplot(1,2,1)
ax1=sns.countplot(x = 'day_of_week', data=city_hotel_df, hue='is_canceled', order=order)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[1,1])
plt.title("Reservation status per day of week for City Hotel", size=10)
plt.xlabel('Day of week')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.ylim(0, 9000)
plt.xticks(rotation=45)

plt.subplot(1,2,2)
ax1=sns.countplot(x = 'day_of_week', data=resort_hotel_df, hue='is_canceled', order=order)
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=[1,1])
plt.title("Reservation status per day of week for Resort Hotel", size=10)
plt.xlabel('Day of week')
plt.ylabel('number of reservations')
plt.legend(['not cancelled','cancelled'])
plt.ylim(0, 9000)
plt.xticks(rotation=45)

plt.show()
```

```python
cancelled_df = df[df['is_canceled']==1]
cancelled_df_adr = cancelled_df.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values(by='reservation_status_date',inplace=True)

not_cancelled_df = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values(by='reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'], label='cancelled')
plt.legend()
plt.show()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr= not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```
